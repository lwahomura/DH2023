```python
test_y = 0

def test():
    # global test_y
    test_y += 1

print(test_y)
test()
print(test_y)
```

```python
test_y = 0

def test():
    global test_y
    test_y += 1

print(test_y)
test()
print(test_y)
```

```python
test_y = 0

def test(val):
    return val + 1

print(test_y)
test_y = test(test_y)
print(test_y)
```

```python
test_y = 0

def test():
    print(test_y)

print(test_y)
test()
print(test_y)
```

Нелокальная

```python
def test():
  test_z = 0
  def test_2():
    nonlocal test_z
    test_z += 1
  test_2()
  print(test_z)

test()
```

```python
def test(a):
  test_w = 2
  def test_2():
    return a + test_w
  return test_2()

test(5)
```

```python
def test(a):
  test_w = 2
  def test_2():
    test_w += a
    return test_w
  return test_2()

test(5)
```

```python
def test_multi(a):
    def test_multi_internal(b):
        return b*a
    return test_multi_internal


test_multi_5 = test_multi(5)
print(test_multi_5(9))
```

Как мы уже смотрели, для каких-то данных есть interned objects, для кого-то их нет

```python
def compare_addrs(a, b):
  print(a is b)

a = -6
b = -6
compare_addrs(a, b)

a = "bad string"
b = "bad string"
compare_addrs(a, b)
```

```python
bad_vals = [-6, "bad string"]
for val in bad_vals:
  a = val
  b = val
  compare_addrs(a, b)

bad_vals = [-6, "bad string"]
for i in range(0, len(bad_vals)):
  a = bad_vals[i]
  b = bad_vals[i]
  compare_addrs(a, b)
```

```python
array_to_cut = [1,1,2,3,4,3,3,5]
for i in range(0,len(array_to_cut)):
  if array_to_cut[i] == 3:
    array_to_cut.pop(i)
print(array_to_cut)
```

```python
array_to_change = [1,2,3,4,5]
for i in range(0,len(array_to_change)):
  if array_to_change[i] == 3:
    array_to_change[i+1] -=1

print(array_to_change)
```
